```python
import matplotlib.pyplot as plt
import numpy as np
from numpy import sin, pi
from IPython.display import Audio
```

```python
t = np.linspace(0, 1, 320)
data = sin(2 * pi * 1 * t) # < - 1 Hz
plt.plot(t, data)
plt.grid(True, which='both')
plt.axhline(y=0, color='k')
```

```python
def play(freq, sec=1):
    framerate = 44100 # < - rate of sampling
    t = np.linspace(0, sec, framerate * sec)
    data = np.sin(2 * pi, freq * t) # < - sine function formula
    return Audio(data, rate=framerate, autoplay=True) # play the generated sound
```

```python
data = np.random.randn(44100)
Audio(data, rate=44100)
```

```python
freq = 110
play(freq) # any base frequency
play(freq * 2)
play(freq * 4)
```
